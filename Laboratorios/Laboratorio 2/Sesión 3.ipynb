## 1.1. Preparaci칩n de los Datos

```python
# Realizamos split de train y test al 80%
sample_df = df.sample(frac=0.025, random_state=19)
train_df, test_df = train_test_split(sample_df, test_size=0.2, random_state=19)
train_df = train_df.reset_index(drop=True)
test_df = test_df.reset_index(drop=True)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=3, sharey=True, figsize=(9, 3))
sns.countplot(x='labels',data=df, ax=ax[0])
ax[0].set_title('Dataset completo')
sns.countplot(x='labels',data=train_df, ax=ax[1])
ax[1].set_title('Conjunto de entrenamiento')
sns.countplot(x='labels',data=test_df, ax=ax[2])
ax[2].set_title('Conjunto de prueba')
plt.show()
```

```python
# Redimensionamos las imagenes a 100x100 en una misma escala para poderlas normalizar
temp = []
for img_name in train_df.image:
    img_path = os.path.join("/content/cat-dog-images-for-classification/cat_dog", img_name)
    img = imread(img_path)
    img = resize(img, (100, 100), anti_aliasing=True)
    temp.append(np.array(img))

train_x = np.stack(temp)
```

```python
# Redimensionamos las imagenes a 100x100 en una misma escala para poderlas normalizar
temp = []
for img_name in test_df.image:
    img_path = os.path.join("/content/cat-dog-images-for-classification/cat_dog", img_name)
    img = imread(img_path)
    img = resize(img, (100, 100), anti_aliasing=True)
    temp.append(np.array(img))

test_x = np.stack(temp)
```

**Normalizar los datos:**

Vamos a normalizar el valor de los pixeles de las im치genes en el rango [0,1]. Al tener una escala uniforme el modelo converge m치s r치pido y puede aumentar el rendimiento al evitar que ciertas caracter칤sticas dominen otras.

```python
x_train = train_x.astype('float32') / 255.0
x_test = test_x.astype('float32') / 255.0
```

# 2. Modelamiento

En esta secci칩n, emplearemos el algoritmo K-Nearest Neighbors (KNN) para nuestro problema de clasificaci칩n de im치genes. KNN es un algoritmo basado en instancia que clasifica una imagen seg칰n la mayor칤a de clases en sus K im치genes vecinas m치s cercanas en el espacio de caracter칤sticas.

En el contexto de este problema (clasificaci칩n de perros y gatos) la simplicidad de KNN nos permite comparar las caracteristicas de las im치genes de una manera intuitiva.

##### Consideraci칩n

Debemos tener en cuenta que KNN no puede procesar im치genes directamente. Las im치genes deben ser transformadas en vectores de caracter칤sticas. Una forma com칰n es aplanar cada imagen en un vector unidimensional. Por ejemplo, una imagen de 100x100x3 se convierte en un vector de 30000 elementos.

```python
# Aplanamiento de las im치genes
x_train_flatten = np.array([img.flatten() for img in x_train])
x_test_flatten = np.array([img.flatten() for img in x_test])
```

```python
# Tama침o del vector unidimensional
print(x_train_flatten.shape)
```

No hemos definido nuestros y_train y y_test

```python
train_df.info()
```

```python
y_train = train_df.labels
y_test = test_df.labels
```

Ahora vamos a desarrollar una cuadr칤cula para encontrar los mejores par치metros posibles a nuestro modelo.

```python
start = time.time()

param_grid = {
    'n_neighbors': np.arange(2, 8, 1),
    'weights': ['uniform'],
    'metric': ['euclidean', 'manhattan']
}

# Crear la instancia de GridSearchCV (Grilla de b칰squeda)
grid_search = GridSearchCV(KNeighborsClassifier(), param_grid, scoring="accuracy", # Nos concentraremos en la exactitud (cu치ntas predicciones son correctas)
                           cv=6, return_train_score=True)

# Entrenar el modelo
grid_search.fit(x_train_flatten, y_train)

end = time.time()

print("Tiempo de entrenamiento: {:.2f} segundos".format(end - start))
```

```python
print("Mejor precisi칩n: ", grid_search.best_score_)
print("Mejores par치metros: ", grid_search.best_params_)
```

```python
unique_train = np.unique(y_train)
print("Matriz de confusi칩n datos de entrenamiento")

# Matriz de confusi칩n para entrenamiento
cm = confusion_matrix(y_train, grid_search.best_estimator_.predict(x_train_flatten))
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=unique_train)
disp.plot()
```

```python
# Mejor modelo
best_model = grid_search.best_estimator_
# Predecir
y_pred = best_model.predict(x_test_flatten)
# Exactitud
print("Exactitud: ", accuracy_score(y_test, y_pred))
```

```python
unique_test = np.unique(y_test)
print("Matriz de confusi칩n datos de test")

# Matriz de confusi칩n para test
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=unique_test)
disp.plot()
```

# 3. PCA

PCA es una t칠cnica de reducci칩n de dimensionalidad. PCA identifica las direcciones en las cuales los datos var칤an m치s y proyecta los datos en un espacio de menor dimensi칩n, reteniendo la mayor cantidad de informaci칩n posible. Al reducir la dimensionalidad, PCA permite que KNN opere en un espacio de caracter칤sticas m치s manejable, mejorando potencialmente la velocidad y el rendimiento del modelo.

PCA requiere que los datos est칠n en un formato bidimensional donde cada fila es una muestra y cada columna es una caracter칤stica. Nuestras im치genes Tienen que ser aplanadas en un vector unidimensional para poder aplicar PCA. por lo que podemos usar X_train_flat y X_test_flat.

```python
# Aplicar PCA
pca = PCA(0.95) # Queremos conservar el 95% de la varianza
x_train_pca = pca.fit_transform(x_train_flatten)
x_test_pca = pca.transform(x_test_flatten)
```

```python
start = time.time()

# Crear la instancia de GridSearchCV (Grilla de b칰squeda)
param_grid = {
    'n_neighbors': np.arange(2, 8, 1),
    'weights': ['uniform'],
    'metric': ['euclidean', 'manhattan']
}

grid_search = GridSearchCV(KNeighborsClassifier(), param_grid, cv=6, return_train_score=True)

grid_search.fit(x_train_pca, y_train)

end = time.time()

print("Tiempo de entrenamiento: {:.2f} segundos".format(end - start))
```

```python
print("Mejor precisi칩n: ", grid_search.best_score_)
print("Mejores par치metros: ", grid_search.best_params_)
```

```python
print("Matriz de confusi칩n datos de entrenamiento")

# Matriz de confusi칩n para entrenamiento
cm = confusion_matrix(y_train, grid_search.best_estimator_.predict(x_train_pca))
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=unique_train)
disp.plot()
```

```python
# Mejor modelo
best_model = grid_search.best_estimator_
# Predecir
y_pred = best_model.predict(x_test_pca)
# Exactitud
print("Exactitud: ", accuracy_score(y_test, y_pred))
```

```python
print("Matriz de confusi칩n datos de test")

# Matriz de confusi칩n para test
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=unique_test)
disp.plot()
```

```python
plt.figure(figsize=(15,15))
for i in range(0, min(len(y_pred), 9)):
    # Seleccionar un 칤ndice aleatorio
    idx = np.random.randint(0, len(y_pred))

    # Mostrar imagen
    plt.subplot(3,3, i+1)
    plt.imshow(x_test[idx].reshape(100,100,3)*255.0, cmap='gray')
    plt.title(f"Predicted {y_pred[idx]}, Actual {y_test[idx]}")
plt.show()
```

La diferencia en las m칠tricas de desempe침o entre el modelo entrenado con datos originales y el modelo entrenado con datos transformados por PCA no es significativa. Sin embargo, se observa una notable reducci칩n en el tiempo de entrenamiento del modelo, pasando de 322.2s a 6.1s. PCA permite conservar la esencia de la informaci칩n en un espacio de menor dimensi칩n, lo cual es beneficioso al trabajar con datos de alta dimensionalidad como las im치genes.
